```python
from pytube import YouTube


url = 'https://www.youtube.com/watch?v=9E6b3swbnWg'
name = 'samples/chopin'

yt = YouTube(url)
stream = yt.streams.get_audio_only()
stream.download(filename=name)
```

```python
# We inport the librosa library

import librosa.display

# IMporting the matplotlib library

import matplotlib.pyplot as plt

# We can use the librosa library to load the audio file

audio, sample_rate = librosa.load('samples/chopin')

# We can use the librosa library to plot the audio file

plt.figure(figsize=(14, 5))
librosa.display.waveshow(audio, sr=sample_rate)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# define the function that you want to animate
def func(x):
    return x**2

# set the x-axis limits for the plot
xmin, xmax = -10, 10

# create the figure and axis objects for the plot
fig, ax = plt.subplots()

# set the y-axis limits based on the function
ymin, ymax = func(xmin), func(xmax)
ax.set_ylim(ymin, ymax)
ax.set_xlim(xmin, xmax)

# create a line object to represent the function
line, = ax.plot([], [], lw=2)

# define the function that will be called at each frame of the animation
def update(num):
    # set the x-coordinates of the line based on the current frame number
    x = np.linspace(xmin, xmax, num+1)
    # set the y-coordinates of the line based on the function
    y = func(x)
    # update the line object with the new x and y coordinates
    line.set_data(x, y)
    return line,

# create the animation object
anim = FuncAnimation(fig, update, frames=range(100), blit=True)

# display the animation in the notebook
HTML(anim.to_jshtml())
```
