```python
# We start creating some toy data for 
# classification and regression problems

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification, make_regression
from sklearn.model_selection import train_test_split

# Create a classification dataset of 4 classes
X, y = make_classification(n_samples=1000, n_features=10, n_classes=4, n_informative=5, n_redundant=5)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# How to score the relevance of a feature
from sklearn.feature_selection import mutual_info_classif
mutual_info_classif(X_train, y_train).round(2)
```

```python
# We plot the second, last and sixth feature in a 3d plot

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_train[:, 1], X_train[:, 5], X_train[:, 9], c=y_train, cmap='rainbow')
ax.set_xlabel('Feature 2')
ax.set_ylabel('Feature 6')
ax.set_zlabel('Feature 10')
plt.show()

```
